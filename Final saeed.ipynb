```python
Image=[]
for i in range(1,5):
    img = cv2.imread(f'image{i}.jpg')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)/255.0
    Image.append(img)
```

```python
f,a = plt.subplots(2,2)
a[0,0].imshow(Image[0], cmap='gray')
a[0,1].imshow(Image[1], cmap='gray')
a[1,0].imshow(Image[2], cmap='gray')
a[1,1].imshow(Image[3], cmap='gray')
plt.show()
```

```python
Images=[]
for i in range(1,5):
    imgs = cv2.imread(f'images{i}.jpg')
    imgs = cv2.cvtColor(imgs, cv2.COLOR_BGR2GRAY)/255.0
    Images.append(imgs)
```

```python
f,a = plt.subplots(2,2)
a[0,0].imshow(Images[0], cmap='gray')
a[0,1].imshow(Images[1], cmap='gray')
a[1,0].imshow(Images[2], cmap='gray')
a[1,1].imshow(Images[3], cmap='gray')
plt.show()
```

```python
for i in range(4):
    width1,height1=Image[i].shape
    width2,height2=Images[i].shape
    area1=width1*height1
    area2=width2*height2
    print(area1)
    print(area2)
    print(Image[i].shape)
    if area1>area2:
        Image[i]=cv2.resize(Image[i], (height2,width2))
        print(Image[i].shape)
    elif area2>area1:
        Images[i]=cv2.resize(Images[i], (height1,width1))
        print(Image[i].shape)
    else:
        continue
```

```python
f,a = plt.subplots(2,2)
a[0,0].imshow(Image[0], cmap='gray')
a[0,1].imshow(Image[1], cmap='gray')
a[1,0].imshow(Image[2], cmap='gray')
a[1,1].imshow(Image[3], cmap='gray')
plt.show()
```

```python
for i in range(4):
    Image[i]=cv2.GaussianBlur(Image[i],(55,55),0)
    Images[i]=Images[i]-cv2.GaussianBlur(Images[i],(85,85),0)
```

```python
f,a = plt.subplots(2,2)
a[0,0].imshow(Image[0], cmap='gray')
a[0,1].imshow(Image[1], cmap='gray')
a[1,0].imshow(Image[2], cmap='gray')
a[1,1].imshow(Image[3], cmap='gray')
plt.show()
```
